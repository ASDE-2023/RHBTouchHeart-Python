## Conditional

- Conditional statement that runs a different set of statements depending on whether an expression is true or false.

- `If elif else` statements are used to store the conditions, and the python code within, will be executed based on if the condition is true or false.

- Operators such as `==`, `!=`, `>`, `<` can be used to manipulate custom made conditions

```python
food = "ice_cream"
if(food == "nasi_lemak"):
  print("Ikram loves this food!!")
else:
  print("Ikram does not like this food...")
```

```python
first_num = 5
second_num = 6
third_num = 7

if first_num < second_num:
  print("First condition triggererd")
elif first_num + third_num == 10:
  print("Second condition triggered")
elif second_num > third_num:
  print("Third condition triggered")
```

## For Loop

- A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

- For loop can also be used to iterate through numbers

- With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python
## example of for loop iterating through numbers
##
for i in range(1,4):
  print(i)
```

```python
## example for loop iterating through a list

items = ["item 1", "item 2", "item 3"]

for item in items:
  print(item)
```

```python
dict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"],
  "color_type":["chrome","matte", "normal"]
}

for item in dict:
  print(item)
```

```python
## Libraries import for data manipulation and visualization
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None)
data = pd.read_csv("titanic.csv")

data.head(5)
```

## Data Manipulation

- Note that the original data is not manipulated, because different changes are stored in different variables for demonstration purposes.

```python
## Dropping columns in the data
df_dropped = data.drop('Survived', axis=1)
df_dropped.head()
```

```python
## Dropping multiple columns in data
df_dropped_multiple = data.drop(['Survived', 'Name'], axis=1)
df_dropped_multiple.head()
```

```python
## renaming columns in a dataset
data.columns
```

```python
df_renamed = data.rename(columns={'PassengerId': 'Id'})
df_renamed.head()
```

```python
## renaming multiple columns
df_renamed_multiple = data.rename(
    columns={
        'PassengerId': 'Id',
        'Sex': 'Gender',
    }
)
df_renamed_multiple.head()
```

```python
## Select column with specific datatype
integer_data = data.select_dtypes('int')
integer_data.head()
```

```python
## Select specific value in the column
data[data['Pclass'] == 1]
```

```python
## Group by
data.groupby('Sex').agg({'PassengerId': 'count'})
```

```python
## Replacing values in a dataframe
data['Sex'].replace(['male', 'female'], ["M", "F"])
```

## Data Visualization

```python
## Categorical visualization
variable = 'Survived' ## change this to different columns

var=data[variable]
#count number of categorical variable-group by
varvalue=var.value_counts()
plt.figure(figsize=(9,3))
plt.bar(varvalue.index, varvalue)
plt.xticks(varvalue.index,varvalue.index.values)
plt.ylabel("Frequency")
plt.title(variable)
plt.show()
```

```python
## Numerical visualization
numericvar=['Fare', 'Age', 'PassengerId']

var=data[numericvar[1]]
plt.figure(figsize=(9,3))
plt.hist(var, bins=50)
plt.xlabel(variable)
plt.ylabel("Freq")
plt.show()
```
